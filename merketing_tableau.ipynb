```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os

# Define the parameters
campaigns = [
    "Buy One Get One Free",
    "Spring Fresh Discounts",
    "Seasonal Joy: Holiday Specials"
]
locations = [
    "Sayajigunj", 
    "Alkapuri", 
    "Gorwa", 
    "Vasna", 
    "Fatehganj", 
    "Gotri", 
    "Manjalpur"
]
ad_types = ["Medium", "Newspapers", "Magazines", "Flyers", "Posters", "Pamphlets"]

# Create a list to hold all the data
data = []

# Generate data for each year from 2014 to 2018
for year in range(2014, 2019):
    for month in range(1, 13):
        # Generate a random date for the month
        num_days = (datetime(year, month + 1, 1) - timedelta(days=1)).day
        day = np.random.randint(1, num_days + 1)
        date = datetime(year, month, day)

        # Randomly select values for other columns
        campaign = np.random.choice(campaigns)
        discount = np.random.randint(10, 51)  # Random discount between 10% and 50%
        location = np.random.choice(locations)
        ad_type = np.random.choice(ad_types)
        number_of_copies = np.random.randint(50, 301)  # Random number of copies
        print_cost = np.random.randint(2000, 10001)  # Random print cost
        footfall = np.random.randint(100, 501)  # Random footfall
        conversion = np.random.randint(1, footfall + 1)  # Random conversion based on footfall
        conversion_rate = conversion / footfall if footfall > 0 else 0
        per_unit_costing = print_cost / number_of_copies if number_of_copies > 0 else 0
        cost = print_cost
        roi = (conversion * discount) - cost  # Basic ROI calculation

        # Append the data
        data.append({
            "Month": date.strftime("%B"),
            "Year": year,
            "Date": date.strftime("%d-%m-%Y"),
            "Campaign": campaign,
            "Discount (%)": discount,
            "Location": location,
            "Type of Ad": ad_type,
            "Number of Copies": number_of_copies,
            "Print Cost": print_cost,
            "Footfall": footfall,
            "Conversion": conversion,
            "Conversion Rate": conversion_rate,
            "Per Unit Costing": per_unit_costing,
            "Cost": cost,
            "ROI": roi
        })

# Specify the output directory
output_dir = "D:/Maitri folder/study/tableau-project-marketing"  # Adjust this path as needed
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Specify the full path for the Excel file
output_file = os.path.join(output_dir, "marketing_data_2014_2018.xlsx")

# Create a DataFrame
df = pd.DataFrame(data)

# Export to Excel
df.to_excel(output_file, index=False)

print(f"Excel file '{output_file}' generated successfully.")
```
